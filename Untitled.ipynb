```python
#this for the tabular files


try:
    businesses_df = pd.read_csv('Businesses.csv')
    print("Businesses.csv loaded successfully:")
    print(businesses_df.head())
    print("-" * 30)
except FileNotFoundError:
    print("Error: Businesses.csv not found.")

try:
    income_df = pd.read_csv('Income.csv')
    print("Income.csv loaded successfully:")
    print(income_df.head())
    print("-" * 30)
except FileNotFoundError:
    print("Error: Income.csv not found.")

try:
    population_df = pd.read_csv('Population.csv')
    print("Population.csv loaded successfully:")
    print(population_df.head())
    print("-" * 30)
except FileNotFoundError:
    print("Error: Population.csv not found.")

# Load the README.md file
try:
    with open('README.md', 'r') as f:
        readme_content = f.read()
    print("README.md loaded successfully:")
    print(readme_content[:200] + "...") # Print the first 200 characters
    print("-" * 30)
except FileNotFoundError:
    print("Error: README.md not found.")

# Load the Stops.txt file using pandas for better handling of delimiters and quotes
try:
    stops_df = pd.read_csv('Stops.txt', sep=',', quotechar='"', keep_default_na=False, na_values=[''])
    print("Stops.txt loaded successfully using pandas:")
    print(stops_df.head())
    print("-" * 30)
except FileNotFoundError:
    print("Error: Stops.txt not found.")
except pd.errors.ParserError as e:
    print(f"Error parsing Stops.txt: {e}")
    print("Please ensure the file is properly comma-separated and handles quotes correctly.")
```

```python
#this is for the images file using gpd
catchments_primary = gpd.read_file("/Users/hadishafat/2001-Group/catchments/catchments_primary.shp")
display(catchments_primary.head())
catchments_secondary = gpd.read_file("/Users/hadishafat/2001-Group/catchments/catchments_secondary.shp")
display(catchments_secondary.head())
catchments_future = gpd.read_file("/Users/hadishafat/2001-Group/catchments/catchments_future.shp")
display(catchments_future.head())



```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))  # Adjust the figure size as needed

# Plot each GeoDataFrame on the same axes
catchments_primary.plot(ax=ax, color='blue', label='Primary', edgecolor='black', linewidth=0.5)
catchments_secondary.plot(ax=ax, color='green', label='Secondary', edgecolor='black', linewidth=0.5)
catchments_future.plot(ax=ax, color='red', label='Future', edgecolor='black', linewidth=0.5)

# Add a legend to distinguish the layers
ax.legend()

# Add a title to the plot
ax.set_title("Combined Catchment Areas")

# Optionally, set the axis limits to a common area if needed.
# You might need to determine these limits based on your data.
# For example:
# xmin, xmax, ymin, ymax = combined_catchments.total_bounds
# ax.set_xlim(xmin, xmax)
# ax.set_ylim(ymin, ymax)

# Show the plot
plt.show()

#  Database operations (these are unchanged, but now happen *after* the combined plot)
#  It's good practice to separate plotting and data persistence.
catchments_primary.to_postgis("catchments_primary", con=engine, if_exists="replace", index=False)
catchments_secondary.to_postgis("catchments_secondary", con=engine, if_exists="replace", index=False)
catchments_future.to_postgis("catchments_future", con=engine, if_exists="replace", index=False)
```
